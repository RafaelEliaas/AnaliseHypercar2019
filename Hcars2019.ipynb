Análise Hyper Carros 2019

Explicação da Base de Dados: A base de dados fornece informações detalhadas sobre uma variedade de carros de alto desempenho, apresentando várias características-chave de cada modelo. 

Bibliotecas Utilizadas:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Código de Carregamento de Dados:

```python
df = pd.read_csv('Hyper Cars 2019.csv')
df.head(n=10)
```

Exploração de arquivo:

```python
df.shape
```

Temos 10 linhas de arquivo divididos em 7 colunas

Informação das colunas:
1 - Car Name: O nome do carro.
2 - Displacement: A capacidade do motor em litros (por exemplo, 6.1 significa 6.1 litros).
3 - Engine: O tipo de motor, como V12, V8, etc.
4 - hp: A potência do motor em cavalos de potência (hp).
5 - Transmission: O tipo de transmissão, representado por um número.
6 - Top-speed: A velocidade máxima do carro em milhas por hora (mph).
7 - Cost: O custo do carro, representado em dólares (USD)

Tipo das Colunas:

```python
df.dtypes
```

Correção dos erros de escrita nos títulos das colunas

```python
df.rename(columns={'     Car Name':'Car Name'},inplace=True)
df.rename(columns={'Displacement ':'Displacement'},inplace=True)
df.rename(columns={'Enginee':'Engine'},inplace=True)
df.rename(columns={'          hp':'HorsePower (HP)'},inplace=True)
df.rename(columns={'   Transmission':'Transmission'},inplace=True)
df.rename(columns={'    Toppp-speeed':'Top-Speed'},inplace=True)
df.rename(columns={'           Cost':'Cost'},inplace=True)
```

```python
df.head(n=10)
```

Correção de erros nos nomes dos carros:

```python
df['Car Name'].replace('McLaren Sennaaaa','McLaren Senna',inplace=True)
df['Car Name'].replace('Koenigseeeggg Agera','Koenigsegg Agera',inplace=True)
df['Car Name'].replace('Pagaaani Huayra BC','Pagani Huayra BC',inplace=True)
df['Car Name'].replace('Bugatti Veyronre','Bugatti Veyron',inplace=True)
```

```python
df.head(n=10)
```

Arredondando os números da Coluna Displacement, para isso devemos converte-la para float, uma vez que a coluna se encontra como object (creio ser por conta do "Hello" na linha 5)

```python
fn = lambda valor: float(valor.replace("Hello","NaN"))
```

```python
df['Displacement'] = df['Displacement'].apply(fn)
```

```python
df.dtypes
```

```python
df['Displacement'] = df['Displacement'].round(2)
```

```python
df.head(n = 10)
```

Correção dos valores da coluna Engine:

```python
df['Engine'].replace('V8987654','V8',inplace=True)
df['Engine'].replace('__','V12',inplace=True)
```

```python
df.head(n = 10)
```

Convertendo o Top Speed de mph (Milhas por Hora) para kmph (Quilometros por Hora). Para isso utilizarei uma função lambda

```python
df['Top-Speed'] = df['Top-Speed'].apply(lambda x: x * 1.60934)
df['Top-Speed'] = df['Top-Speed'].round(2)
```

```python
df.head(n=10)
```

Visualização de Dados:

Distribuição dos tipos de motor:

```python
motor_counts = df['Engine'].value_counts()

plt.figure(figsize=(9, 4))
motor_counts.plot(kind='bar', color='skyblue')

plt.title('Distribuição dos Tipos de Motor')
plt.xlabel('Tipo de Motor')
plt.ylabel('Número de Carros')

plt.xticks(rotation=0) 
plt.tight_layout()  
plt.show()
```

Valores dos Carros:

```python
dados = df.dropna(subset=['Car Name', 'Cost'])
df = df.sort_values(by='Cost')

plt.figure(figsize=(9, 4))
plt.scatter(dados['Car Name'], dados['Cost'], color='black', alpha=0.6)
plt.title('Gráfico dos Valores dos Carros')
plt.xlabel('Carro')
plt.ylabel('Preço')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Gráfico do Carro e sua Velocidade Máxima

```python
dados = df.dropna(subset=['Car Name', 'Top-Speed'])
df = df.sort_values(by='Top-Speed', ascending=True)


plt.figure(figsize=(7, 4))
plt.scatter(dados['Car Name'], dados['Top-Speed'], color='black', alpha=0.6)
plt.title('Relação entre os Carros e sua Velocidade Máxima')
plt.xlabel('Carro')
plt.ylabel('Top-Speed')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Relação entre o tipo de motor e sua cavalaria:

```python
dados = df.dropna(subset=['Engine', 'HorsePower (HP)'])

dados['Engine'] = dados['Engine'].astype(str)

plt.figure(figsize=(7, 4))
plt.scatter(dados['Engine'], dados['HorsePower (HP)'], color='black', alpha=0.6)
plt.title('Relação entre Tipo de Motor e Cavalos de Potência')
plt.xlabel('Tipo de Motor')
plt.ylabel('Potência (hp)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Média de cavalaria por tipo de motor:

```python
dados = dados.dropna(subset=['Engine', 'HorsePower (HP)'])

media_hp_por_engine = dados.groupby('Engine')['HorsePower (HP)'].mean().sort_values()

plt.figure(figsize=(9, 4))
media_hp_por_engine.plot(kind='bar', color='skyblue')
plt.title('Média de hp por Tipo de Motor')
plt.xlabel('Tipo de Motor')
plt.ylabel('Média de Potência (hp)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Velocidade Máxima por Tipo de Motor

```python
dados = dados.dropna(subset=['Engine', 'Top-Speed'])


top_speed_engine = dados.groupby('Engine')['Top-Speed'].mean().sort_values()

plt.figure(figsize=(9, 4))
top_speed_engine.plot(kind='bar', color='skyblue')
plt.title('Top Speed por Tipo de Motor')
plt.xlabel('Tipo de Motor')
plt.ylabel('Top Speed (km/h)')

# Mostrando o gráfico
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout()  # Ajusta o layout para evitar que os rótulos se sobreponham
plt.show()
```

Velocidade Máxima por Tipo de Transmissão

```python
dados = dados.dropna(subset=['Top-Speed', 'Transmission'])

top_speed_transmission = dados.groupby('Top-Speed')['Transmission'].mean().sort_values()

plt.figure(figsize=(9, 4))
top_speed_transmission.plot(kind='bar', color='skyblue')

plt.title('Top Speed por Tipo de Transmissão')
plt.xlabel('Top Speed (km/h)')
plt.ylabel('Tipo de Transmissão')

plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

Cavalos de Potência por tipo de Transmissão

```python
dados = dados.dropna(subset=['Transmission', 'HorsePower (HP)'])

hp_transmission = dados.groupby('Transmission')['HorsePower (HP)'].mean().sort_values()

plt.figure(figsize=(9, 4))
hp_transmission.plot(kind='bar', color='skyblue')

plt.title('Cavalos de Potência por Tipo de Transmissão')
plt.xlabel('Tipo de Transmissão')
plt.ylabel('Cavalos de Potência (HP)')

plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

Cavalos de Potência em relação a Velocidade Máxima:

```python
dados = dados.dropna(subset=['Top-Speed', 'HorsePower (HP)'])

hp_transmission = dados.groupby('Top-Speed')['HorsePower (HP)'].mean().sort_values()

plt.figure(figsize=(9, 4))
hp_transmission.plot(kind='bar', color='skyblue')

plt.title('Relação entre Cavalos de Potência e Velocidade Máxima')
plt.xlabel('Velocidade Máxima')
plt.ylabel('Cavalos de Potência (HP)')

plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```
