```python
import pandas as pd

df = pd.read_csv('Hyper Cars 2019.csv')
df.head(n=10)
```

Exploração de arquivo:

```python
df.shape
```

Temos 10 linhas de arquivo divididos em 7 colunas

Informação das colunas:
1 - Car Name: O nome do carro.
2 - Displacement: A capacidade do motor em litros (por exemplo, 6.1 significa 6.1 litros).
3 - Engine: O tipo de motor, como V12, V8, etc.
4 - hp: A potência do motor em cavalos de potência (hp).
5 - Transmission: O tipo de transmissão, representado por um número.
6 - Top-speed: A velocidade máxima do carro em milhas por hora (mph).
7 - Cost: O custo do carro, representado em dólares (USD)

Tipo das Colunas:

```python
df.dtypes
```

Correção dos erros de escrita nos títulos das colunas

```python
df.rename(columns={'     Car Name':'Car Name'},inplace=True)
df.rename(columns={'Displacement ':'Displacement'},inplace=True)
df.rename(columns={'Enginee':'Engine'},inplace=True)
df.rename(columns={'          hp':'HorsePower (HP)'},inplace=True)
df.rename(columns={'   Transmission':'Transmission'},inplace=True)
df.rename(columns={'    Toppp-speeed':'Top-Speed'},inplace=True)
df.rename(columns={'           Cost':'Cost'},inplace=True)
```

```python
df.head(n=10)
```

Após corrigir os Nomes das Colunas, iremos corrigir erros nos nomes dos carros:

```python
df['Car Name'].replace('McLaren Sennaaaa','McLaren Senna',inplace=True)
df['Car Name'].replace('Koenigseeeggg Agera','Koenigsegg Agera',inplace=True)
df['Car Name'].replace('Pagaaani Huayra BC','Pagani Huayra BC',inplace=True)
df['Car Name'].replace('Bugatti Veyronre','Bugatti Veyron',inplace=True)
```

```python
df.head(n=10)
```

Arredondando os números da Coluna Displacement, para isso devemos converte-la para float, uma vez que a coluna se encontra como object (creio ser por conta do "Hello" no na linha 5)

```python
fn = lambda valor: float(valor.replace("Hello","NaN"))
```

```python
df['Displacement'] = df['Displacement'].apply(fn)
```

```python
df.dtypes
```

```python
df['Displacement'] = df['Displacement'].round(2)
```

```python
df.head(n = 10)
```

Correção dos valores da coluna Engine:

```python
df['Engine'].replace('V8987654','V8',inplace=True)
df['Engine'].replace('__','V12',inplace=True)
```

```python
df.head(n = 10)
```

Convertendo o Top Speed de mph (Milhas por Hora) para kmph (Quilometros por Hora). Para isso utilizarei uma função lambda

```python
df['Top-Speed'] = df['Top-Speed'].apply(lambda x: x * 1.60934)
df['Top-Speed'] = df['Top-Speed'].round(2)
```

```python
df.head(n=10)
```

Começo da Visualização de Dados:

```python
import matplotlib.pyplot as plt

contagem_engine = df['Engine'].value_counts()

plt.figure(figsize=(7, 3))
contagem_engine.plot(kind='bar', color='skyblue')
plt.title('Distribuição dos Tipos de Motor')
plt.xlabel('Engine')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
dados = df.dropna(subset=['Engine', 'HorsePower (HP)'])

dados['Engine'] = dados['Engine'].astype(str)

plt.figure(figsize=(7, 4))
plt.scatter(dados['Engine'], dados['HorsePower (HP)'], color='black', alpha=0.6)
plt.title('Relação entre Engine e hp')
plt.xlabel('Tipo de Motor')
plt.ylabel('Potência (hp)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
dados = dados.dropna(subset=['Engine', 'HorsePower (HP)'])

# Agrupando os dados por tipo de motor e calculando a média de hp
media_hp_por_engine = dados.groupby('Engine')['HorsePower (HP)'].mean().sort_values()

# Criando o gráfico de barras
plt.figure(figsize=(7, 4))
media_hp_por_engine.plot(kind='bar', color='skyblue')
plt.title('Média de hp por Tipo de Motor')
plt.xlabel('Tipo de Motor')
plt.ylabel('Média de Potência (hp)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
